#### plt sample

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 2*pi with a step of 0.1
x = np.arange(0, 2*np.pi, 0.1)

# Calculate the corresponding y values using the sine function
y = np.sin(x)

axe, fig = plt.subplots(1, 2, figsize=(16, 3))
fig[0].plot(x, y)
fig[0].set_title('Sine function')
fig[0].set_xlabel('x')
fig[0].set_ylabel('y')
fig[0].grid(True)

fig[1].plot(x, y)
fig[1].set_title('Sine function')
fig[1].set_xlabel('x')
fig[1].set_ylabel('y')
fig[1].grid(True)
plt.show()
```
